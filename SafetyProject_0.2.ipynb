```python
import pandas as pd
from sklearn import svm, metrics
from sklearn.model_selection import train_test_split, GridSearchCV
import datetime
import xlrd
from sklearn.ensemble import RandomForestClassifier
```

```python
datas=pd.read_csv('/Users/heesang/ML/datas.csv')
label=pd.read_csv('/Users/heesang/ML/label.csv')
datas
```

```python
#svc.SVC Model
train_data, test_data, train_label, test_label = train_test_split(datas, label)

clf=svm.SVC()
clf.fit(train_data, train_label)

pre=clf.predict(test_data)
score=metrics.accuracy_score(test_label, pre)

print('Score:', score)
```

```python
#RandomForestClassifier
train_data, test_data, train_label, test_label = train_test_split(datas, label)

clf=RandomForestClassifier()
clf.fit(train_data,train_label)
pre=clf.predict(test_data)

score=metrics.accuracy_score(test_label, pre)
print("ì •ë‹µë¥ :", score)
```

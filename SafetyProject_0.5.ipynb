```python
#Init and load previous version data csv files
import pandas as pd
from sklearn import svm, metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import datetime
import xlrd, csv

path='./'
label=[]
datas=[]

#load datas
f = open(path + 'datas_ver4.csv')
reader = csv.reader(f)
for data in reader:
    for i in range(len(data)):
        data[i] = float(data[i])
    datas.append(data)
f.close() 

#load label
f=open(path+'label_ver4.csv')
reader=csv.reader(f)
for lb in reader:
    for i in range(len(lb)):
        lb[i] = int(lb[i])
    label=lb
f.close()
```

```python
fire_df=pd.read_excel(path + "서울시 화재 출동현황(2018.1_2019.11).xlsx")
fire_df
```

```python
#pandas Timestamp.weekday returns mon==0, sunday==6

for idx, row in fire_df.iterrows():
    datas[idx].append(row['출동일시'].weekday())
```

```python
datas
```

```python
#svc.SVC Model
train_data, test_data, train_label, test_label = train_test_split(datas, label)

clf=svm.SVC()
clf.fit(train_data, train_label)

pre=clf.predict(test_data)
score=metrics.accuracy_score(test_label, pre)

print('Score:', score)
```

```python
#RandomForestClassifier
train_data, test_data, train_label, test_label = train_test_split(datas, label)

clf=RandomForestClassifier()
clf.fit(train_data,train_label)
pre=clf.predict(test_data)

score=metrics.accuracy_score(test_label, pre)
print("정답률:", score)
```

```python
#monday predict
monday_data=[]
monday_label=[]

for idx in range(len(datas)):
    if (datas[idx][-1] == 6):
        monday_data.append(datas[idx])
        monday_label.append(label[idx])
        
#svc.SVC Model
train_data, test_data, train_label, test_label = train_test_split(monday_data, monday_label)

clf=svm.SVC()
clf.fit(train_data, train_label)

pre=clf.predict(test_data)
score=metrics.accuracy_score(test_label, pre)

print('Score:', score)
```

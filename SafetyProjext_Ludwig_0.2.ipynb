```python
import pandas as pd
from sklearn import svm, metrics
from sklearn.model_selection import train_test_split
import datetime
import xlrd

path='./'
fire_xlsx=pd.read_excel(path+'서울시 화재 출동현황(2018.1_2019.11).xlsx')

# '시도' 칼럼 삭제
fire_xlsx.drop('시도', axis=1, inplace=True)

display(fire_xlsx)
```

```python
# 소방서 이름만 남기기

FirestationParent = []
for idx, row in fire_xlsx.iterrows():
    Parent = "".join(str(row['소방서'][0:-3]))
    FirestationParent.append(Parent)

fire_xlsx.loc[:, 'Parent'] = FirestationParent
display(fire_xlsx)
```

```python
# 안전센터 이름만 남기기

FirestationChild = []
for idx, row in fire_xlsx.iterrows():
    Child = "".join(str(row['119안전센터'][0:-7]))
    FirestationChild.append(Child)

fire_xlsx.loc[:, 'Child'] = FirestationChild
display(fire_xlsx)
```

```python
# '출동요일' 칼럼(Day) 추가 # 0:월요일 ~ 6:일요일
Day = []
for idx, row in fire_xlsx.iterrows():
    weekday = (row['출동일시'].weekday())
    Day.append(weekday)

fire_xlsx.loc[:, 'Day'] = Day
display(fire_xlsx)
```

```python
# '출동시간대' 칼럼(TimeSpan) 추가
import datetime

TimeSpan = []
for idx, row in fire_xlsx.iterrows():
    time = row['출동일시']        
    timeTuple = time.timetuple()
    TimeSpan.append(timeTuple.tm_hour)

fire_xlsx.loc[:, 'TimeSpan'] = TimeSpan
display(fire_xlsx)
```

```python
# 출동 거리 칼럼 추가(distance)
distance=[]

for idx, row in fire_xlsx.iterrows():
    distanceM = int(row['소방서거리(㎞)']*1000)
    distance.append(distanceM)

fire_xlsx.loc[:, 'DistanceMeter'] = distance # index = fire_xlsx.index
display(fire_xlsx)
```

```python
# '소방서', '119안전센터', '시군구', '읍면동', '접수일시', '출동일시', '도착일시' 칼럼 삭제
fire_xlsx.drop("소방서", axis=1, inplace=True)
fire_xlsx.drop("119안전센터", axis=1, inplace=True)
fire_xlsx.drop("접수일시", axis=1, inplace=True)
fire_xlsx.drop("출동일시", axis=1, inplace=True)
fire_xlsx.drop("도착일시", axis=1, inplace=True)
fire_xlsx.drop("시군구", axis=1, inplace=True)
fire_xlsx.drop("읍면동", axis=1, inplace=True)
fire_xlsx.drop("소방서거리(㎞)", axis=1, inplace=True)

display(fire_xlsx)
```
